```python
from sklearn import datasets 
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import cluster
from sklearn.cluster import KMeans
import csv
labelencoder = LabelEncoder()

plt.figure(figsize=(16,12))
arr = pd.read_csv("D:/google download/players.csv")

y = arr[['ID']].values.reshape(-1, 1)


abb = arr.drop('ID', axis = 1)
ab = pd.DataFrame(abb,columns = ["AVG","BIP%"])
x1 = arr.drop('BIP%', axis = 1).drop('ID', axis = 1)
x2 = arr.drop('AVG', axis = 1).drop('ID', axis = 1)
x1 = np.array(x1)
x2 = np.array(x2)
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ["r", "g", "y"]
markers = ['o', 'o', 'o']


# KMeans algorithm 
K = 3
kmeans= KMeans(n_clusters=K).fit(X)

centers = np.array(kmeans.cluster_centers_)
plt.plot()

for i, l in enumerate(kmeans.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')

plt.scatter(centers[:,0], centers[:,1], marker="v", color='orange',s = 120)

plt.show()

idd = np.arange(1,104)

csvfile = open('D:/google download/test4.csv','w',newline='')
writer = csv.writer(csvfile)
writer.writerow(['ID',' Grouping'])
for i in range(0,103):
    writer.writerow([idd[i], kmeans.labels_[i]])
csvfile.close()
```
